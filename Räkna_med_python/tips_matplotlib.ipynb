# Några typer av diagram

- linjediagram - vanliga plt.plot(x,y)
- cirkeldiagram - plt.pie()
- stapeldiagram - plt.bar()
- horisontell stabeldiagram - plt.barh()

```python
import matplotlib.pyplot as plt
import numpy as np 

kön = [8, 26]
labels = ["Flickor", "Pojkar"]

plt.pie(kön, labels = labels, autopct="%1.1f%%")
plt.title("Könsfördelning i en teknikklass")
plt.show()
```
