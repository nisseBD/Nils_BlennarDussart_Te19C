```python
# 1 a)

import pandas as pd

df = pd.read_csv("sysselsattning.csv", encoding="ISO-8859-1", header=1)

förvärv = df[df["sysselsättning"] == "förvärvsarbetande"]

summa = förvärv["2019"].sum()

print(f"{summa} Personer förvärvsarbetade i Sverige år 2019 i ålder 16-44 år")
```

```python
# 1 b)

import pandas as pd

df = pd.read_csv("sysselsattning.csv", encoding="ISO-8859-1", header=1)

förvärv = df[df["sysselsättning"] == "ej förvärvsarbetande"]

summa = förvärv["2019"].sum()

print(f"{summa} Personer förvärvsarbetade inte i Sverige år 2019 i ålder 16-44 år")
```

```python
# 1 c)

import matplotlib.pyplot as plt 
import numpy as np 
import pandas as pd

df = pd.read_csv("sysselsattning.csv", encoding="ISO-8859-1", header=1)

ejförvärv = df[df["sysselsättning"] == "ej förvärvsarbetande"]
förvärv = df[df["sysselsättning"] == "förvärvsarbetande"]

summa1 = förvärv["2019"].sum()
summa2 = ejförvärv["2019"].sum()

personer = [summa1, summa2]
labels = ["Förvärvsarbetande", "Ej förvärvsarbetande"]

plt.pie(personer, labels = labels, autopct="%1.1f%%")
plt.title("Fördelning mellan förvärvsarbetande och icke förvärvsarbetande år 2019")
plt.show()
```

```python
# 3 män

import matplotlib.pyplot as plt 
import numpy as np 
import pandas as pd

df = pd.read_csv("sysselsattning.csv", encoding="ISO-8859-1", header=1)

män = df.loc[(df["kön"] == "män") & (df["sysselsättning"] == "förvärvsarbetande")]
y = män["2019"]

x = män["ålder"]

plt.plot(x, y, "-*")
plt.title("Graf över förvärvsarbetande män genom åldersgrupperna")
plt.xlabel("Ålder")
plt.ylabel("Antal Förvärvsarbetande")
plt.grid()
plt.show()
```
