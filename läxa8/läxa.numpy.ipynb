```python
import matplotlib.pyplot as plt 
import numpy as np 

x = np.arange(1,11)

def f(x):
    y = 3*x+1
    return y

plt.plot(x,f(x))
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```
